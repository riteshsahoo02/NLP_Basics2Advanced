```python
ticket_text = """
PNR: 2456712345
Train No: 12345
Train Name: Rajdhani Express
From: New Delhi
To: Howrah
Departure: 14 Apr 2026 16:55
Arrival: 15 Apr 2026 09:45
Class: 3A
Booking Status: WL 12
Current Status: CNF B2 21
"""
def extract_field(pattern, text):
    match = re.search(pattern, text, re.IGNORECASE)
    return match.group(1).strip() if match else None

data = {
    "pnr" : extract_field(r"PNR\s*[:\-]?\s*(\d{10})", ticket_text),
    "train_no": extract_field(r"Train\s*No\s*[:\-]?\s*(\d{5})", ticket_text),
    "train_name": extract_field(r"Train\s*Name\s*[:\-]?\s*([A-Za-z0-9 &\-]+)",ticket_text),
    "from":extract_field(r"From\s*[:\-]?\s*([A-Za-z ]+)",ticket_text),
    "to":extract_field(r"To\s*[:\-]?\s*([A-Za-z ]+)",ticket_text),
    "departure": extract_field(r"Departure\s*[:\-]?\s*([0-9]{1,2}\s+[A-Za-z]{3}\s+[0-9]{4}\s+[0-9]{2}:[0-9]{2})", ticket_text),
    "arrival": extract_field(r"Arrival\s*[:\-]?\s*([0-9]{1,2}\s+[A-Za-z]{3}\s+[0-9]{4}\s+[0-9]{2}:[0-9]{2})", ticket_text),
    "class": extract_field(r"Class\s*[:\-]?\s*([A-Za-z0-9]+)", ticket_text),
    "booking_status" : extract_field(r"Booking\s*Status\s*[:\-]?\s*([A-Za-z0-9 ]+)", ticket_text),
    "current_status": extract_field(r"Current\s*Status\s*[:\-]?\s*([A-Za-z0-9 ]+)", ticket_text)
}
print(data)
```

```python
import re

def clean_text(text):
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n+", "\n", text)
    return text.strip()

def extract_field(patterns, text):
    for pattern in patterns:
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            return match.group(1).strip()
    return None

def extract_ticket_info(text):
    text = clean_text(text)

    patterns = {
        "pnr": [
            r"PNR(?:\s*No\.?)?\s*[:\-]?\s*(\d{10})"
        ],
        "train_no": [
            r"Train\s*(?:No|Number)\s*[:\-]?\s*(\d{5})"
        ],
        "train_name": [
            r"Train\s*Name\s*[:\-]?\s*([A-Za-z0-9 .&()-]+)"
        ],
        "from": [
            r"From(?:\s*Station)?\s*[:\-]?\s*([A-Za-z .()]+)"
        ],
        "to": [
            r"To(?:\s*Station)?\s*[:\-]?\s*([A-Za-z .()]+)"
        ],
        "departure": [
            r"Departure(?:\s*Time)?\s*[:\-]?\s*([0-9]{1,2}\s+[A-Za-z]{3}\s+[0-9]{4}\s+[0-9]{2}:[0-9]{2})"
        ],
        "arrival": [
            r"Arrival(?:\s*Time)?\s*[:\-]?\s*([0-9]{1,2}\s+[A-Za-z]{3}\s+[0-9]{4}\s+[0-9]{2}:[0-9]{2})"
        ],
        "class": [
            r"Class\s*[:\-]?\s*([A-Za-z0-9]+)"
        ],
        "booking_status": [
            r"Booking\s*Status\s*[:\-]?\s*([A-Za-z0-9/ ]+)"
        ],
        "current_status": [
            r"Current\s*Status\s*[:\-]?\s*([A-Za-z0-9/ ]+)"
        ]
    }

    extracted = {}
    for field, field_patterns in patterns.items():
        extracted[field] = extract_field(field_patterns, text)

    return extracted


ticket_text = """
PNR No. : 2456712345
Train Number-12345
Train Name : Rajdhani Express
From Station : New Delhi
To Station : Howrah
Departure Time : 14 Apr 2026 16:55
Arrival Time : 15 Apr 2026 09:45
Class : 3A
Booking Status : WL/12
Current Status : CNF/B2/21
"""

data = extract_ticket_info(ticket_text)
print(data)
```

```python
def print_ticket_summary(data):
    print("\n===== RAIL TICKET SUMMARY =====")
    print(f"PNR No.          : {data.get('pnr')}")
    print(f"Train No.        : {data.get('train_no')}")
    print(f"Train Name       : {data.get('train_name')}")
    print(f"From             : {data.get('from')}")
    print(f"To               : {data.get('to')}")
    print(f"Departure        : {data.get('departure')}")
    print(f"Arrival          : {data.get('arrival')}")
    print(f"Class            : {data.get('class')}")
    print(f"Booking Status   : {data.get('booking_status')}")
    print(f"Current Status   : {data.get('current_status')}")
    print("================================")
print_ticket_summary(data)
```

## Stemming 

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
words = words = ["running", "runs", "ran", "studies", "better"]

print("Stemming:")
for word in words:
    print(word,":- ",stemmer.stem(word))
print("\nLemmatization:")
for w in words:
    print(w,":- ",lemmatizer.lemmatize(w,pos='v'))
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]","",text)
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    print("Stemming:")
    for token in tokens:
        print(token,": ",stemmer.stem(token))
    print("\nLemmatization:")
    for token in tokens:
        print(token,":",lemmatizer.lemmatize(token, pos='v'))

text = "He is running and studying better than before"
preprocess(text)
```

## BOW (Bag Of Words)
    Bag of Words is a technique to convert text into numerical vectors by counting the frequency of words in a document. It ignore grammar and order of the words in the sentence. It only check which words appear and how many times it appears.

```python
text = ["I love One Piece","I love NLP","I love AI"]
text = [sent.lower() for sent in text]
print(text)
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(text)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
## include bigrams and unigrams
vectorizer = CountVectorizer(ngram_range=(1,2))
X = vectorizer.fit_transform(text)

print(vectorizer.get_feature_names_out())
```

### Q. What is ngram_range?
    It controls the number of consecutive words considered as features. For example, (1,2) includes unigrams and bigrams.

```python
text = ["I love machine learning and deep learning"]
vectorizer = CountVectorizer(ngram_range=(2,3))
X = vectorizer.fit_transform(text)

print(vectorizer.get_feature_names_out())
```

## Tf-Idf
    TF-IDF measures how important a word is in a document relative to the entire corpus. A word is important if it appears frequently in a document and rarely in other documents.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

sentences = [
    "the movie is good",
    "the movie is bad",
    "machine learning is powerful",
    "machine learning is easy"
]

vectorizer = TfidfVectorizer(stop_words='english')

X = vectorizer.fit_transform(sentences)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

## Text Classification

```python
texts = [
    "Win a free iPhone now",
    "Limited offer click now",
    "Meeting at 5 pm",
    "Let's have lunch tomorrow",
    "Congratulations you won a prize",
    "Project deadline is tomorrow"
]

labels = [
    "spam",
    "spam",
    "not spam",
    "not spam",
    "spam",
    "not spam"
]
labels = [1 if l == "spam" else 0 for l in labels]

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X, labels)

test = ["Free prize waiting for you"]
test_vec = vectorizer.transform(test)
prediction = model.predict(test_vec)
print("Spam" if prediction[0] == 1 else "Not Spam")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, labels)

test = ["Free prize waiting for you"]
test_vec = vectorizer.transform(test)
prediction = model.predict(test_vec)
print("Spam" if prediction[0] == 1 else "Not Spam")
```
